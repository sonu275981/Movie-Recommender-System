we have to convert above genres and also keywords,cast,crew list in '[Action','Adventure','Fantasy'] form and remove key from dictonary

for this we will create helper function

we have to convert string of list into list form by "literal_eval" in "ast" function

```python
import ast
def convert(obj):
    keyy = []
    for i in ast.literal_eval(obj):
        keyy.append(i['name'])
    return keyy    
```

```python
movie_pro['genres'] = movie_pro['genres'].apply(convert)
movie_pro.head(5)
```

Repeat same process with keywords

```python
movie_pro['keywords'] = movie_pro['keywords'].apply(convert)
movie_pro.head(2)
```

for cast column we want top 3 actor of the movie

```python
def convert1(obj):
    keyy = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            
            keyy.append(i['name'])
            counter+=1
        else:
            break
    return keyy  
```

```python
movie_pro['cast'] = movie_pro['cast'].apply(convert1)
movie_pro.head(2)
```

checking for movie director only

```python
def fetch_director(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i["job"]=='Director':
            director.append(i['name'])
            break
    return director  
```

```python
movie_pro['crew'] = movie_pro['crew'].apply(fetch_director)
movie_pro.head(2)
```

converting "overview" column into list

```python
movie_pro['overview'] = movie_pro['overview'].apply(lambda x:x.split())
movie_pro.head(2)
```

Removing spaces between names

```python
movie_pro['genres'] =  movie_pro['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie_pro['keywords'] =  movie_pro['keywords'].apply(lambda x:[i.replace("","") for i in x])
movie_pro['cast'] =  movie_pro['cast'].apply(lambda x:[i.replace("","") for i in x])
movie_pro['crew'] =  movie_pro['crew'].apply(lambda x:[i.replace("","") for i in x])
movie_pro.head(3)
```

```python
movie_pro['tags'] = movie_pro['overview'] + movie_pro['genres'] + movie_pro['keywords'] + movie_pro['cast'] + movie_pro['crew']
movie_pro.head(3)
```

```python
movie_df = movie_pro[['movie_id','title','tags']]
movie_df.head(3)
```

Now again convert back "Tag" column into string

```python
movie_df['tags'] = movie_df['tags'].apply(lambda x:" ".join(x))
movie_df['tags'] = movie_df['tags'].apply(lambda x:x.lower())
```

```python
movie_df['tags'][0]
```

first we collect most frequent 5000 words from tags column without any stopword and then check with each 5000 movies for which words are frequenty used in certin movie and than by this we make dataframe of 5000 rows and 5000 column 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
pss = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(pss.stem(i))
    return " ".join(y)    
```

```python
movie_df['tags'] = movie_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
movie_vectors = cv.fit_transform(movie_df['tags']).toarray()
```

Now we know we have already convert each movie into a vector so we will find distence between each movie for recommendation

Here we will not find Euclidean distance instead of this we will find cosine distance

Euclidean is tip to tip distance b/w two point

cosine is angle b/w two point 

*** for high dimensioanl data Euclidean distance is not good/reliable so we can use cosine distance

```python
# we can use cosine distance or cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

coss_similarity = cosine_similarity(movie_vectors)
```

Now recommending Top nearest movie

we want sorting on basis of 2nd no. in the list

```python
def recommend(movie):
    moive_index = movie_df[movie_df['title'] == movie].index[0]
    distences = coss_similarity[moive_index]
    movie_list = sorted(list(enumerate(distences)), reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(movie_df.iloc[i[0]].title)
    
    return
```

```python
recommend("Batman Begins")
```
